```python
%matplotlib inline
from pylab import *
import numpy as np 

np.random.seed(2)

pageSpeed = np.random.normal(3.0, 1.0, 100)
purchaseAmount = np.random.normal(50.0,30.0,100)/pageSpeed

scatter(pageSpeed,purchaseAmount)
```

```python
trainX = pageSpeed[:80]
trainY = purchaseAmount[:80]

testX = pageSpeed[80:]
testY = purchaseAmount[80:]
```

```python
scatter(trainX, trainY)
```

```python
scatter(testX,testY)
```

```python
x = np.array(trainX)
y = np.array(trainY)

p4 = np.poly1d(np.polyfit(x,y,6))
```

```python
import matplotlib.pyplot as plt

xp = np.linspace(0, 7, 100)
axes = plt.axes()
axes.set_xlim([0,7])
axes.set_ylim([0, 200])
plt.scatter(x, y)
plt.plot(xp, p4(xp), c='r')
plt.show()
```

```python
testx = np.array(testX)
testy = np.array(testY)

axes = plt.axes()
axes.set_xlim([0,7])
axes.set_ylim([0, 200])
plt.scatter(testx, testy)
plt.plot(xp, p4(xp), c='r')
plt.show()
```

```python
from sklearn.metrics import r2_score
r2 = r2_score(testy,p4(testx))
print(r2)
```

```python
from sklearn.metrics import r2_score
r2 = r2_score(trainY, p4(trainX))
print(r2)
```
